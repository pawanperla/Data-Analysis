#**Cleaning Data - Casting Datatypes and Handling Missing Values**







```python
import numpy as np

people2 = {
    'first': ['Pawan', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Perla', 'Doe', 'Doe', 'Woakes', np.nan, np.nan, 'Missing'],
    'email': ['pawn@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['19', '55', '63', '36', None, None, 'Missing']
}

dic3 = pd.DataFrame(people2)
dic3
```

```python
dic3.replace("NA",np.nan,inplace=True)
dic3.replace("Missing",np.nan,inplace=True)
dic3
```

```python
dic3.dropna(axis="index",how='all',subset=["email","last"])
```

```python
dic3.isna()
```

```python
dic3.dtypes
```

```python
dic3['age'] = dic3['age'].astype(float)
dic3['age'].mean()
```
